```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=11)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=11)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = (df_train['class'] == 'f').astype('int').values
y_val = (df_val['class'] == 'f').astype('int').values
y_test = (df_test['class'] == 'f').astype('int').values
```

```python
import matplotlib.pyplot as plt

max_depths = range(1, 30)  # Profundidades de 1 a 12
train_aucs = []
val_aucs = []

for max_depth in max_depths:
    dt = DecisionTreeClassifier(max_depth=max_depth)
    dt.fit(X_train, y_train)

    # AUC para el conjunto de entrenamiento
    y_pred_train = dt.predict_proba(X_train)[:, 1]
    train_auc = roc_auc_score(y_train, y_pred_train)
    train_aucs.append(train_auc)

    # AUC para el conjunto de validaciÃ³n
    y_pred_val = dt.predict_proba(X_val)[:, 1]
    val_auc = roc_auc_score(y_val, y_pred_val)
    val_aucs.append(val_auc)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(max_depths, train_aucs, label='AUC Train', marker='o')
plt.plot(max_depths, val_aucs, label='AUC Val', marker='o')
plt.title('AUC en funciÃ³n de la profundidad del Ã¡rbol')
plt.xlabel('Profundidad mÃ¡xima del Ã¡rbol')
plt.ylabel('AUC')
plt.xticks(max_depths)
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, det_curve
import seaborn as sns

dt = DecisionTreeClassifier(max_depth=9)
dt.fit(X_train, y_train)

# Realizar predicciones
y_pred = dt.predict(X_val)

# Crear la matriz de confusiÃ³n
cm = confusion_matrix(y_val, y_pred)

# Definir las etiquetas
labels = ['Perfil Real (0)', 'Perfil Falso (1)']

# Graficar la matriz de confusiÃ³n
plt.figure(figsize=(3, 3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=labels, yticklabels=labels)

# Personalizar el grÃ¡fico
plt.title('Matriz de ConfusiÃ³n')
plt.xlabel('PredicciÃ³n')
plt.ylabel('Realidad')
plt.show()
```

```python
# Realizar predicciones y obtener probabilidades
y_scores = dt.predict_proba(X_val)[:, 1]

# Obtener FPR y TPR
fpr, tpr, thresholds = roc_curve(y_val, y_scores, pos_label=1)

# Calcular FNR
fnr = 1 - tpr  # Tasa de Falsos Negativos

# Graficar la curva DET
plt.figure(figsize=(8, 6))
plt.plot(fpr, fnr, marker='o', color='blue', label='Curva DET')
plt.title('Curva DET (Detection Error Tradeoff)')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Falsos Negativos (FNR)')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="--")

# AÃ±adir lÃ­nea de referencia (error igual)
plt.plot([0.001, 1], [0.001, 1], color='red', linestyle='--', label='Error Igual')

plt.legend()
plt.show()
```
